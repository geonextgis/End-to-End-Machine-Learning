## **Data Preprocessing**

### **One-Hot Encoding (OHE) using Pandas**

One-hot encoding is a technique used in data preprocessing, specifically in the context of categorical variables, to convert them into a binary format that can be more easily used for machine learning algorithms. Categorical variables are variables that represent categories or groups and don't have a natural numerical order. One-hot encoding is essential for algorithms like Logistic Regression, Support Vector Machines, and Neural Networks that require numerical inputs.

```python
# Checking the unique values of each column
df["specialisation"].value_counts()
```

```python
# Creating a list of columns for which we need dummy variables
columnList = ["gender", "ssc_board", "hsc_board", "hsc_subject", 
              "undergrad_degree", "work_experience", "specialisation"]
```

```python
# Creating the dummy variables
dummies = pd.get_dummies(df, columns=columnList, drop_first=True)
dummies.head()
```

```python
# Convert 'Placed' into 1 and 'Not Place' into 0 in status column
# Create a mapping dictionary
statusDict = {"Placed": 1, "Not Placed": 0}
# Mapping the dictionary on the status column
dummies["status"] = dummies["status"].map(statusDict)
dummies.head()
```

```python
# Checking the dummy dataframe information
dummies.info()
```

```python
# Changing the order of columns
dummies.columns
```

```python
# Creating a list of column order
column_order = ['ssc_percentage', 'hsc_percentage', 'degree_percentage',
       'emp_test_percentage', 'mba_percent', 'gender_M',
       'ssc_board_Others', 'hsc_board_Others', 'hsc_subject_Commerce',
       'hsc_subject_Science', 'undergrad_degree_Others',
       'undergrad_degree_Sci&Tech', 'work_experience_Yes',
       'specialisation_Mkt&HR', 'status']
```

```python
# Create the final dataframe
final = dummies[column_order]
final.head()
```

```python
# Checking the final dataframe
final.info()
```

## **Exploratory Data Analysis (EDA)**

```python
# Create a pair plot
sns.pairplot(df, hue="status")
plt.show()
```

## **Extract the Independent and Dependent Columns**

```python
# Extracting the independent columns (x)
x = final.iloc[:, 0:14]
x.head()
```

```python
# Extracting the dependent columns (y)
y = final.iloc[:, -1]
y.head()
```

## **Perform Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=30)
```

```python
x_train
```

```python
y_train
```

```python
x_test.head()
```

```python
y_test.head()
```

## **Train the Model**

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Creating an object from LogisticRegression class
classifier = LogisticRegression()
```

```python
# Training the model
classifier.fit(x_train, y_train)
```

## **Validate the Model**

```python
# Predicted the test data using trained classifier
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
